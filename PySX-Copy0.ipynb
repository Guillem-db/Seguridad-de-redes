### <a name="5.1">5.1</a> Procesado y visualización

A continuación cargaremos datos estadísticos del retardo para distintos puertos de una misma IP. Suponemos que en un canal de transmisión ideal para un Host inactivo el retardo obtenido será el mismo para todos los puertos, mientras que si el host esta online es probable que obtengamos tiempos de respuesta diferentes en función del puerto escaneado.

Otra forma de detectar disponibilidad de hosts será obteniend valores de respuesta muy bajos (Linux y móviles) o inusualmente alto (Windows). 

```python
delays=scanner.load_delays()
scanner.resources.head()
```

```python
def zscore(x):
    return (x-x.mean())/x.std()

def plot_normalized(scanner):
    m=scanner.resources['max'].copy()
    df=scanner.resources[['min','std']].copy()
    df_plot=df.apply(lambda x: x/m.values)
    df_plot['max']=m/m.max()
    return df_plot
df=plot_normalized(scanner)
```

Para comparar los distintos valores de retardo hemos normalizado los valores máximos, mínimos y desviaciones estandar para determinar que hosts estan online. A continuación se muestran los resultados obtenidos al escanear una red wifi de mala calidad en busca de Hosts activos:

```python
df[['min','max','std']].plot(kind='bar',figsize=(20,10),subplots='true',title='Vista general de los retardos')
```

Tal y como se observa a primera vista obtenemos unos valores de std mucho más altos de lo esperado debido a las malas condiciones del canal inalámbrico, por lo que nuestra hipótesis de bajas *std* para hosts inactivos queda parcialmente invalidada: **En escaneos a través de conexiones con jitter una baja *std* es condición suficiente pero no necesaria para que un host esté inactivo.**

Al agrupar los valores de retardo más pequeños es posible distinguir inmediatamente tres hosts activos:

```python
delays.sort('min',ascending=True)['min'].head(10).plot(kind='bar',figsize=(10,5),title='Menores tiempos de respuesta')
```

Analizar de forma análoga los mayores tiempos de respuesta nos permite descubrir dos nuevos hosts activos.

```python
delays.sort('max',ascending=False)['max'].head(10).plot(kind='bar',figsize=(10,5),title='Mayores tiempos de respuesta')
```

En cambio, al analizar los valores de desviación estándar de los retardos a través de wifi podemos comprobar como el jitter se carga cualquier información potencialmente útil para distinguir hosts activos. En este caso una desviación estándar muy baja nos indicaría que el host no está activo, mientras que una *std* alta no nos aportaria información, ya que podría deberse a que el objetivo contesta peticiones a diferentes puertos con distinto retardo o que el canal inalambrico introduce jitter.

Para corregir este efecto, y aunque no está implementado, en teoría podría hacerse más de una petición a un mismo puerto. En un canal ideal dos peticiones iguales al mismo puerto deberían tardar los mismo en responderse, por lo que si se obtuviera una *std* alta podría asociarse al jitter del canal. 

```python
delays.sort('std',ascending=False)['std'].head(20).plot(kind='bar',figsize=(10,5),title='Mayores desviaciones tipicas en tiempos de respuesta')
```

```python
%%file Delays.py
import numpy as np
import pandas as pd


def df_resource_delays_from_result(sr,classifier):
    """Obtiene un DataFrame que contiene la informacion del escaner procesada"""
    mi,ma,std,mean=ip_delays_from_results(sr)
    delays=ip_delay(mi[0],ma[0],std[0],mean[0],classifier)
    for i in range(1,len(mi)):
        delays=delays.combine_first(ip_delay(mi[i],ma[i],std[i],mean[i],classifier))
    return delays 

def ips_from_results(sr):
    """sr: DataFrame que contiene los resultados de un escaneo
       retorna una lista que contiene las ips de las que hemos obtenido respuesta en el escaner"""
    return list(set(sr[sr.delay>0].address))

def ip_delays_from_results(sr):
    """Obtiene las diferentes metricas derivadas de los retardos"""
    return [sr[sr.delay>0][sr.address==ip].min() for ip in ips_from_results(sr)],[sr[sr.delay>0][sr.address==ip].max() for ip in ips_from_results(sr)],[sr[sr.delay>0][sr.address==ip].std() for ip in ips_from_results(sr)],[sr[sr.delay>0][sr.address==ip].mean() for ip in ips_from_results(sr)]

def ip_delay(mi,ma,std,mean,classifier=None,maxdelay=10000,mindelay=1000):
    """con las metricas derivadas del delay determina si el recurso esta disponible o no"""  
    delays=pd.DataFrame(index=[mi.address],columns=['ip','max','min','mean','std','alive'])
    delays.ip=mi.address
    delays['max']=ma.delay
    delays['min']=mi.delay
    delays['mean']=mean.delay  
    delays['std']=std.delay  
    delays['mdmax']=delays['min']/delays['max']
    delays['maxdm']=delays['max']/delays['min']  
    delays['maxsm']=delays['max']-delays['min'] 
    if classifier:  
      delays['alive']=classify_delay(delays,classifier)  #AI
    else:      
      delays['alive']=False  
      if delays['max'].values>maxdelay:
        delays['alive']=False
      elif delays['min'].values<mindelay:
          delays['alive']=True    
    return delays  
def classify_delay(delays,classifier):
    X=delays.drop(['ip','alive'],axis=1)
    #print "pred",clf.predict(X)
    return classifier.predict(X)
```
